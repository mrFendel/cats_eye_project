```python
# Функция преобразования json в txt, разнесение фалов по необходимым папкам
maps = {'Eye': 0}
def write_txt(files, to):
    c = 0
    for i, f in tqdm.tqdm_notebook (enumerate(files)):
        try:
            file = open(f, 'r')
            j = json.load(file)
            cl = maps[j['objects'][0]['classTitle']]
            # print(j['objects'][0]['points']['exterior'])
            file.close()
            points = j['objects'][0]['points']['exterior']
            x1 = points[0][0]
            y1 = points[0][1]
            x2 = points[1][0]
            y2 = points[1][1]

            height = j['size']['height']
            width = j['size']['width']

            x0 = (x2 - (x2 - x1) / 2) / width
            y0 = (y2 - (y2 - y1) / 2) / height

            w = (x2 - x1) / width
            h = (y2 - y1) / height
            c+=1

            name = f.replace('.jpg.json', '.txt').split('/')[-1]

            if c % 5 == 0:
    #             with open(to + 'val/' + name, 'w', encoding='utf-8') as tofile:
    #             # tofile = open(to + '/' + name, 'w')

    #                 tofile.write("{} {} {} {} {}".format(cl, x0, y0, w, h))

    #             shutil.copy('../img/' + f.replace('.json', ''), main_dir + '/data/images/val/' + f.replace('.json', ''))
    #         else:
              with open(to + '/labels/val/' + name, 'w', encoding='utf-8') as tofile:
              # tofile = open(to + '/' + name, 'w')

                  tofile.write("{} {} {} {} {}".format(cl, x0, y0, w, h))
            else:
    #             with open(to + 'val/' + name, 'w', encoding='utf-8') as tofile:
    #             # tofile = open(to + '/' + name, 'w')

    #                 tofile.write("{} {} {} {} {}".format(cl, x0, y0, w, h))

    #             shutil.copy('../img/' + f.replace('.json', ''), main_dir + '/data/images/val/' + f.replace('.json', ''))
    #         else:
              with open(to + '/labels/train/' + name, 'w', encoding='utf-8') as tofile:
              # tofile = open(to + '/' + name, 'w')

                  tofile.write("{} {} {} {} {}".format(cl, x0, y0, w, h))

            # src=f.replace('/ann','/img').replace('.json','')
            # dst=to+'images/'+where
            # shutil.copy(src,dst)

        except IndexError:
            print(j,f)
```
